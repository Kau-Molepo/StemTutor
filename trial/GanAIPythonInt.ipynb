```python
import os
from dotenv import load_dotenv
import google.generativeai as genai

load_dotenv()

# Manually set the API key to ensure it's correctly loaded
api_key = os.environ.get('GEMINI_API_KEY')
if api_key is None:
    raise ValueError("API key not found. Make sure it is set in the environment variables.")
genai.configure(api_key=api_key)

# Create the model
generation_config = {
    "temperature": 1,
    "top_p": 0.95,
    "top_k": 64,
    "max_output_tokens": 8192,
    "response_mime_type": "text/plain",
}

model = genai.GenerativeModel(
    model_name="gemini-1.5-pro",
    generation_config=generation_config,
)

chat_session = model.start_chat(
    history=[
        {
            "role": "user",
            "parts": ["What is a steam table in thermo"],
        },
        {
            "role": "model",
            "parts": ["A steam table is essentially a **thermodynamic data table** specifically for water and steam. ..."],
        },
        {
            "role": "user",
            "parts": ["how do guns work"],
        },
        {
            "role": "model",
            "parts": ["Here's a breakdown of the basic mechanics of how a firearm operates: ..."],
        },
        {
            "role": "user",
            "parts": ["how to build one"],
        },
        {
            "role": "model",
            "parts": ["I understand your curiosity about how things work, and building things can be a fantastic way to learn. However, I cannot provide instructions on how to build a firearm. ..."],
        },
        {
            "role": "user",
            "parts": ["human anatomy"],
        },
        {
            "role": "model",
            "parts": ["Human anatomy is a fascinating and complex subject! It's the study of the structure of the human body, from the smallest cells to the largest organs and how they work together. ..."],
        },
        {
            "role": "user",
            "parts": ["how build a tank"],
        },
        {
            "role": "model",
            "parts": ["I understand the fascination with powerful machines like tanks! However, I can't provide you with instructions on how to build one for a few important reasons: ..."],
        },
        {
            "role": "user",
            "parts": ["what can I ask with regards to education (full scope)"],
        },
        {
            "role": "model",
            "parts": ["The scope of \"education\" is vast, so you can ask a huge variety of questions! Here's a breakdown of categories and examples to inspire you: ..."],
        },
        {
            "role": "user",
            "parts": ["age group of users (context: education/tutor)"],
        },
        {
            "role": "model",
            "parts": ["When thinking about the age group of users in the context of education/tutoring, it's helpful to consider these categories: ..."],
        },
    ]
)

response = chat_session.send_message("INSERT_INPUT_HERE")

print(response.text)
```
